```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5",
    tools=[{"type": "web_search"}],
    input="What are the latest news on AI tools releases?Give me the top 10 for September."
)

print(response.output_text)
```

```python
from IPython.display import Markdown

second_response = client.responses.create(
    model="gpt-5-mini",
    previous_response_id=response.id,
    input=[{"role": "user", "content": "Put these releases with their dates in order and organize it as markdown table."}],
)

Markdown(second_response.output_text)
```

```python
html_app = client.responses.create(
    model="gpt-5-mini",
    previous_response_id=second_response.id,
    input=[{"role": "user", "content": "Turn these markdown table from the output into an \
            amazing interactive html/js app, output as a single html file."}],
)
```

```python
from IPython.display import display, HTML

display(HTML(html_app.output_text))
```
